```python
df = df.loc[df["entity"]=="participant"].dropna(how="all",axis=1).drop_duplicates().reset_index(drop=True).filter(['name', #'primary_key_type', 'coding_name', 'concept', 'units',
       'description', 'folder_path', 'title'],axis=1)

print(df.shape[0])
```

```python
## drop future instances (keep instance 0 ando ther stuff)
df = df.loc[~df["title"].str.contains("Instance [1-9]")]
df = df.loc[~df["title"].str.contains("Source of report",case=False)] ## icd related - "Source of report of Q97" - not interesting
df = df.loc[~df["title"].str.contains("device id",case=False)]
df = df.loc[~df["title"].str.contains("method ",case=False)] ## don't remove "methods", remove "method". +- "assesment center"
df = df.loc[~df["title"].str.contains("Reason for skipping",case=False)]
df = df.loc[~df["title"].str.contains("displayed|pilot",case=False)] # duration if X displayed. And things that were in the pilot study

df[1] = df['folder_path'].str.split(">",expand=True)[0].str.strip()
df[2] = df['folder_path'].str.split(">",expand=True)[1].str.strip()
df[3] = df['folder_path'].str.split(">",expand=True)[2].str.strip()

df["title_0"] = df["title"].str.split("|",expand=True)[0]
```

```python
drop_cat1_list =["Biological samples"]
drop_cat2_list = ["Imaging","Biological sampling","Procedural metrics",
                 'Genotypes', 'Telomeres', 'Imputation', 'Whole genome sequences','Exome sequences',
                 "Coronavirus COVID-19",  # could keep one subcat3 of it, drop for now
                  "Consent",
                 ]

print(df.shape[0])
print(df.drop_duplicates(["title_0"]).shape[0])
df = df.loc[~df[1].isin(drop_cat1_list)]
df = df.loc[~df[2].isin(drop_cat2_list)]
print(df.shape[0])
print(df.drop_duplicates(["title_0"]).shape[0])

df
```
